```python
#Since PRICE2 has all NULL value so drop column PRICE2
df1=df.drop(['PRICE2'],axis = 1) #axis=1 indicate the column
df1
```

```python
#Drop row with all NULL value column
df2=df1.dropna(how='all')
df2
```

```python
#Drop row that contain any of the NULL value at the column
df3=df2.dropna(how='any')
df3
```

STEP5: Handle Missing Value

```python
# Importing the class called SimpleImputer from impute model in sklearn
from sklearn.impute import SimpleImputer

#Using df4 as copy of df1
df4=df1.copy()
#To replace the missing value we create below object of SImpleImputer class
imputer = SimpleImputer (missing_values=np.nan, strategy = 'mean')
imputer.fit(df4.iloc[:,:]) #since it is all numerical column

#Replace missing value using the transform method
df4.iloc[:,:] = imputer.transform(df4.iloc[:,:])
```

```python
#Checking how many missing value in each column after removing the NULL value
df4.isnull().sum()
```
