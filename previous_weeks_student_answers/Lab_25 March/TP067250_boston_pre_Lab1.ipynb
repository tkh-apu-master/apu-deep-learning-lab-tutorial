### Data Cleaning

```python
df1 = df.drop(['PRICE2'], axis = 1)
```

```python
null = df1[df1.isnull().any(axis=1)]
null
```

```python
df1 = df1.dropna(how = 'all')
```

```python
df1
```

```python
df1.info()
```

```python
df2 = df1.copy()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(df2.iloc[:, 0:13])

df2.iloc[:, 0:13] = imputer.transform(df2.iloc[:, 0:13])
```
