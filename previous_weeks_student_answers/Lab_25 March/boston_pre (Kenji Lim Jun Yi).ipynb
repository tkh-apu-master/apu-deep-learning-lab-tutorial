```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#import dataset
path = '/content/drive/MyDrive/Colab Notebooks/boston_pre.csv'
df = pd.read_csv(path)
df
```

```python

# check missing values
df.isnull().sum()
```

```python
#drop rows where all cells is N/A
df1=df.dropna(how='all')
df1
```

```python
df2 =df1.drop(['PRICE2'],axis =1)
df2
```

```python
# Importing the classs called SimpleImputer from impute model in sklearn
from sklearn.impute import SimpleImputer
# To replace the missing value we create  below object of SimpleImputer class
imputer = SimpleImputer(missing_values=np.nan, strategy ='mean')
imputer.fit(df2.iloc[:,0:13])

#Replacing the missing value using transform model
df2.iloc[:,0:13] = imputer.transform(df2.iloc[:,0:13])
df2
```
