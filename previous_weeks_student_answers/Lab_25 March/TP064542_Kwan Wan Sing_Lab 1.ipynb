#Read the dataset & Import libraries

```python
#Mounting our Drive to Google Colab
from google.colab import drive
drive.mount('/drive')

# Importing the dataset
import numpy as np
import pandas as pd

path = '/drive/My Drive/Colab Notebooks/boston_pre.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

#Check Missing value

```python
# check for missing values
print(df.isnull().sum())
```

#Handle missing value

```python
#Drop Nan column
df = df.drop(columns=['PRICE2'])
df.info
```

```python
df.head()
```

#Impute Missing value

```python
# Importing the class caleed SimpleImputer from impute model in sklearn
from sklearn.impute import SimpleImputer

# To replace the missing value we create below object of SimpleImputer class
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df.iloc[:, :])

#Replacing the missing value using transfrom method
df.iloc[:, :] = imputer.transform(df.iloc[:, :])

# check for missing values after imputation
print(df.isnull().sum())
```
