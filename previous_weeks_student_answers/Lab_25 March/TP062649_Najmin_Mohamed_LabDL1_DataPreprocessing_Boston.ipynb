**Step 3: Removing Blank Columns & Rows**

```python
# Drop blank column
df1 = df.drop(['PRICE2'], axis = 1)
```

```python
# Drop blank row
df2 = df1.dropna(how='all')
```

**Step 4: Handling Missing Values**

```python
df2.isnull().sum()
```

```python
# Importing the SimpleImputer class from impute model in sklearn
from sklearn.impute import SimpleImputer

# To replace the missing values we create below object of SimpleImputer Class
imputer = SimpleImputer(missing_values=np.nan, strategy = 'mean')
imputer.fit(df2.iloc[:,0:13])
```

```python
# Replacing the missing value using the transform method
df2.iloc[:,0:13] = imputer.transform(df2.iloc[:,0:13])
```

**Step 6: Splitting Training & Testing Data**

```python
# Target variable
y = df2['PRICE']

# Independent variables
X = df2.drop('PRICE', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```
