```python
#Importing the libraries
import numpy as np
import pandas as pd
```

```python
path = '/drive/MyDrive/Colab Notebooks/boston_pre.csv'

df = pd.read_csv(path)
```

```python
df
```

```python
df1 = df.dropna(axis = 1, how='all') #No rows dropped. 
```

```python
df1
```

```python
df1.isnull().sum() #Check missing value
```

```python
df1.shape
```

```python
#Impute data
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean') 
imputer.fit(df1.iloc[:,:])

df1.iloc[:,:] = imputer.transform(df1.iloc[:,:])
```

```python
df1.isnull().sum() #Check missing value again (no more NA)
```
