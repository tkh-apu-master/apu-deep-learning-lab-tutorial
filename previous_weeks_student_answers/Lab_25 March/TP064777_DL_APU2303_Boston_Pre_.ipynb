Step 4: Taking Care of Missing Data

```python
#Total Missing Value for each variable/ feature/ column
df.isnull().sum()
```

```python
#Drop rows where all cells in that row is NA
df1= df.dropna(how= 'all')
df1
```

```python
df1= df.drop (['PRICE2'], axis = 1)
df1
```

```python
df1= df1.dropna(how= 'any') #use this careful as this will remove all the rows which containing NA value
df1
```

Step 5: Missing Values

```python
# Importing the class called SimpleImputer from impute model sklearn
from sklearn.impute import SimpleImputer

#To replace the missing value we create below object of SimpleImputer class
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df1.iloc[:, 1:15]) #indicating column 1 and 2, column 3 is excluded, age and salary included

#Replacing the missing value using transform method
df1.iloc[:, 1:15] =imputer.transform(df1.iloc[:, 1:15]) #iloc [:, 1:15], the : is selecting all row, if we want 1 to 3 row, then we can put 1:4
```
