```python
# To remove price2 column since price2 contains only NA
df1 = df.dropna (axis = 1, how='all')
df1
```

```python
# Drop rows where all cells in that row is NA
df2 = df1.dropna(how='all')
df2
```

```python
# since age contains extreme values (age min 3, max 100), to use median

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer.fit(df2.iloc[:, 6:7]) 

df2.iloc[:, 6:7] = imputer.transform(df2.iloc[:, 6:7])
```

```python
# other to replace with mean since no extreme values

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df2.iloc[:, 0:14]) 

df2.iloc[:, 0:14] = imputer.transform(df2.iloc[:, 0:14])
```
