# Step 5: Categorical Features

```python
df2['CHAS'].value_counts()
```

```python
df2 = df2.astype({'CHAS':'int'})
df2.head()
```

```python
# one-hot encoding
chas_dummies = pd.get_dummies(df2['CHAS'])
chas_dummies.rename(columns = {0:'CHAS_0',1:'CHAS_1'}, inplace=True)
chas_dummies
```

```python
df2 = pd.concat([df2, chas_dummies], axis=1)
df2.drop(['CHAS'], axis=1, inplace=True)
df2
```
