```python
#Copy a completely new dataframe
df1 = df.copy()
```

```python
#Drop "PRICE2" column
#According to the result generated with "info()" and "df.isnall().sum()" function, "PRICE2" column shows all rows are null
df1 = df1.drop(['PRICE2'], axis = 1) 
df1
```

```python
#Drop "rows" having missing values **"all" columns values are NA
df2 = df1.dropna(how="all")
df2
```

```python
#Drop any "rows" with missing values **any NA values are present
df3 = df1.dropna(how="any")
df3
```

```python
#Importing the class called SimpleImputer from impute model in sklearn
from sklearn.impute import SimpleImputer

#To replace the missing value we create below object of SimpleImputer class
imputer = SimpleImputer(missing_values=np.nan,  strategy="mean")
imputer.fit(df1.iloc[:, 1:3])

#Replacing the missing value using transform method
df1.iloc[:, 1:3] = imputer.transform(df1.iloc[:, 1:3])

df1
```
