#**Step1: Import the libraries**

```python
#Importing libraries
import numpy as np      
import pandas as pd
```

#**Step2: Import the dataset**

```python
path = '/drive/MyDrive/Colab Notebooks/boston_pre.csv'
df = pd.read_csv(path)
```

#**Step3: Taking care of missing data**

```python
df.isnull().sum()
```

```python
df1 = df.dropna(how='all')
df1 = df1.drop(['PRICE2'], axis = 1)
df1
```

```python
df1.isnull().sum()
```

#**Step4: Imputing missing values**

```python
#Importing the class callwed SimpleImputer from impute model in sklearn
from sklearn.impute import SimpleImputer

#To replace the missing value we create below object of SimpleImputer class\
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df1.iloc[:, 0:13])

#Replacing the missing value using transform method
df1.iloc[:,0:13] = imputer.transform(df1.iloc[:,0:13])
```
