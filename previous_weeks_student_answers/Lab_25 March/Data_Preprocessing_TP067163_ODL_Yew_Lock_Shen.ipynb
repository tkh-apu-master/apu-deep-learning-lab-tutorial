**Step 4: Remove blank column and row**

```python
# drop blank row
df1 = df.dropna(how='all')
df1
```

```python
# drop blank column
df2 = df1.drop(['PRICE2'],axis=1)
```

```python
# drop blank row
df3 = df2.dropna(how='all')
```

**Step 5: Handle Missing Value**

```python
# Check for Total missing values for each feature/column
df3.isnull().sum()
```

```python
# Importing the class called SimpleImputer from impute model is sklearn
from sklearn.impute import SimpleImputer

# To replace the missing values we create below object of SimpleImputer Class
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df3.iloc[:,0:14])  

#Replacing the missing value using transform method
df3.iloc[:,0:14] = imputer.transform(df3.iloc[:,0:14])
```

```python
# After this, check again if all missing value being imputed
df3.isnull().sum()
```
