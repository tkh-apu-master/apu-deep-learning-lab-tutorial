```python
from keras.datasets import boston_housing

# load the dataset
(X_train, y_train), (X_test, y_test) = boston_housing.load_data()
```

```python
print("x_train shape", X_train.shape)  
print("y_train shape", y_train.shape)
print("x_test shape", X_test.shape)
print("y_test shape", y_test.shape)
```

```python
# finding mean & standard deviation for train data
X_train_mean = X_train.mean(axis=0)
X_train_std = X_train.std(axis=0)
```

```python
# Standardise the data to help with training
X_train -= X_train_mean
X_train /= X_train_std
X_test -= X_train_mean
X_test /= X_train_std
```

```python
# Import required library
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import initializers
from keras import regularizers
from keras.callbacks import EarlyStopping
```

```python
# building a linear stack of layers with the sequential model

L1=200
L2=100

model = Sequential()
model.add(Dense(L1,input_dim=13,activation='relu')) #for first hidden layer
model.add(Dense(L2,activation='relu')) #for second hidden layer

# output later
model.add(Dense(1,activation='linear')) 
```

```python
# compiling the sequential model
model.compile(loss='mse', metrics=['mae'], optimizer='rmsprop')
```

```python
# training the model for 10 epochs
history = model.fit(X_train,y_train, epochs=100,
                    batch_size=100, validation_data=(X_test, y_test))
```

```python
import matplotlib.pyplot as plt

acc2 =history.history['mae']
val_acc = history.history['val_mae']
loss2 = history.history['loss']
val_loss = history.history['val_loss'] #validation loss
epochs = range(1,len(acc2)+1)


plt.plot(epochs, acc2,'b',label='Train MAE')
plt.plot(epochs, val_acc,'r',label='Test MAE')
plt.legend()
plt.figure()


plt.plot(epochs, loss2,'b',label='Train Loss')
plt.plot(epochs, val_loss,'r',label='Test Loss')
plt.legend()
plt.show()
```

Use dropout regularization to prevent overfitting:

```python
# building a linear stack of layers with the sequential model

L1=200
L2=100

model = Sequential()
model.add(Dense(L1,input_dim=13,activation='relu',kernel_initializer='he_normal',
                kernel_regularizer=regularizers.l2(0.01))) #for first hidden layer
model.add(Dense(L2,activation='relu')) #for second hidden layer
model.add(Dropout(0.25))

# output later
model.add(Dense(1,activation='linear')) 


# compiling the sequential model
model.compile(loss='mse', metrics=['mae'], optimizer='rmsprop')

# training the model for 100 epochs
history = model.fit(X_train,y_train, epochs=100,
                    batch_size=100, validation_data=(X_test, y_test))

import matplotlib.pyplot as plt

acc2 =history.history['mae']
val_acc = history.history['val_mae']
loss2 = history.history['loss']
val_loss = history.history['val_loss'] #validation loss
epochs = range(1,len(acc2)+1)


plt.plot(epochs, acc2,'b',label='Train MAE')
plt.plot(epochs, val_acc,'r',label='Test MAE')
plt.legend()
plt.figure()


plt.plot(epochs, loss2,'b',label='Train Loss')
plt.plot(epochs, val_loss,'r',label='Test Loss')
plt.legend()
plt.show()

```

Use Early Stopping

```python
# building a linear stack of layers with the sequential model

L1=200
L2=100

model = Sequential()
model.add(Dense(L1,input_dim=13,activation='relu',kernel_initializer='he_normal',
                kernel_regularizer=regularizers.l2(0.01))) #for first hidden layer
model.add(Dense(L2,activation='relu')) #for second hidden layer
model.add(Dropout(0.25))

# output later
model.add(Dense(1,activation='linear')) 


# compiling the sequential model
model.compile(loss='mse', metrics=['mae'], optimizer='rmsprop')

early_stopping_monitor = EarlyStopping(patience=3)

# training the model for 100 epochs
history = model.fit(X_train,y_train, epochs=100,
                    batch_size=100, validation_data=(X_test, y_test),
                    callbacks=[early_stopping_monitor])

import matplotlib.pyplot as plt

acc2 =history.history['mae']
val_acc = history.history['val_mae']
loss2 = history.history['loss']
val_loss = history.history['val_loss'] #validation loss
epochs = range(1,len(acc2)+1)


plt.plot(epochs, acc2,'b',label='Train MAE')
plt.plot(epochs, val_acc,'r',label='Test MAE')
plt.legend()
plt.figure()


plt.plot(epochs, loss2,'b',label='Train Loss')
plt.plot(epochs, val_loss,'r',label='Test Loss')
plt.legend()
plt.show()
```

# Grid search

```python
#Building a linear stack of layers with the sequantial model

L1=200
L2=100

def create_model():
  model4 = Sequential()
  model4.add(Dense(L1, input_dim = 13, activation = 'relu'))#first hidden layer (need to add input dimension)
  model4.add(Dense(L2, activation= 'relu'))# second hidden layer (do not need to add input dimension))

  #output layer
  model4.add(Dense(1, activation= 'linear'))

  #Compiling the sequantial model
  model4.compile(loss = 'mse', metrics = ['mae'], optimizer = 'rmsprop')

  return model4
```

```python
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier 
from scikeras.wrappers import KerasRegressor 
```

```python
#create model

model4 = KerasRegressor(model = create_model, loss = 'mse',verbose = 0)

#define the grid search parameters
batch_size = [100, 200]
epochs = [10,15,20]
param_grid =  dict(batch_size =  batch_size,
                   epochs = epochs)
grid = GridSearchCV(estimator = model4,
                    param_grid = param_grid,
                    n_jobs = -1, cv = 3) #n_jobs is processing, cv is cross_validation 
grid_result = grid.fit(X_train, y_train)
```

```python
print("Best parameters: ", grid_result.best_params_)
print("Best score: ", -grid_result.best_score_)
```

# Final Model

```python
# building a linear stack of layers with the sequential model

L1=200
L2=100
L3=50

model = Sequential()
model.add(Dense(L1,input_dim=13,activation='relu',kernel_initializer='he_normal',
                kernel_regularizer=regularizers.l2(0.01))) #for first hidden layer
model.add(Dense(L2,activation='relu')) #for second hidden layer
model.add(Dense(L3,activation='relu'))
model.add(Dropout(0.50))

# output later
model.add(Dense(1,activation='linear')) 


# compiling the sequential model
model.compile(loss='mse', metrics=['mae'], optimizer='Nadam')

# training the model for 100 epochs
history = model.fit(X_train,y_train, epochs=20,
                    batch_size=100, validation_data=(X_test, y_test))

import matplotlib.pyplot as plt

acc2 =history.history['mae']
val_acc = history.history['val_mae']
loss2 = history.history['loss']
val_loss = history.history['val_loss'] #validation loss
epochs = range(1,len(acc2)+1)


plt.plot(epochs, acc2,'b',label='Train MAE')
plt.plot(epochs, val_acc,'r',label='Test MAE')
plt.legend()
plt.figure()


plt.plot(epochs, loss2,'b',label='Train Loss')
plt.plot(epochs, val_loss,'r',label='Test Loss')
plt.legend()
plt.show()
```
