```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Dataset for classification

cancer = load_breast_cancer()
cancer
```

```python
cancer.data.shape
```

```python
# Load data into data frame

data = pd.DataFrame(cancer.data, columns=cancer.feature_names)

# T means Transpose (invert)
data.head().T
```

```python
# Know the column names

data.columns
```

```python
cancer.target
```

```python
data['Target'] = cancer.target
```

```python
data.head()
```

```python
# Split data

# Add Target column into dataframe

# data['Target'] = cancer.target

data.head().T
```

```python
# Split here

X = data.drop('Target', axis = 1)
Y = data['Target']
```

```python
Y.value_counts()
```

```python
# Make training and testing datasets for X and Y.

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 0)

print("X_train shape", X_train.shape)

print("X_test shape", X_test.shape)

print("Y_train shape", Y_train.shape)

print("Y_test shape", Y_test.shape)
```

```python
# Normalisation

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

fit_transform: To get mean and XXX from. Only used in Train dataset

transform: XXX

```python
# Import the Sequential
from keras.models import Sequential
from keras.layers import Dense
```

Note: You're using linear equation, so you're using Dense.

```python
# Build the sequential model

# Number of neurals
L1 = 100 # First hidden layer neurons
L2 = 100 # Second hidden layer neurons

# This means parameters in input model.
input = 30

# Activation functions
activation_function = 'relu'
activation_function2 = 'relu'
output_activation_function = 'sigmoid'

# Initialize object
model = Sequential()

# First hidden layer
model.add(Dense(L1, input_dim=input, activation=activation_function))

# Second hidden layer
model.add(Dense(L2, activation=activation_function2))

# Output layer
model.add(Dense(1, activation=output_activation_function))

```

```python
# Compiling the sequential model
# Default optimiser RMSprop
# Note: You can try change the optimizer and number of epochs to see different results.

model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
# Training the model for 10 epochs
# Check overfitting or underfitting, so lecturer will start with 10, then gradually increase.
# There are Training loss in every epoch
# val means validation
# High loss means overfitting.(Accuracy keep hitting 100%)
# NN may not work very well in certain datasets.
# Note: You CANNOT RUN THIS MULTIPLE TIMES. Later your graph looks unusual. If such things happened, Execute Run All button in Colab again.


history = model.fit(X_train, Y_train, epochs = 10, batch_size = 100, validation_data = (X_test, Y_test))
```

```python
# Visualisations

import matplotlib.pyplot as plt

acc2 = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss2 = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1,len(acc2)+1)

# 'b' and 'r' are blue and red color

plt.plot(epochs, acc2, 'b', label='Train Accuracy')
plt.plot(epochs, val_acc, 'r', label='Test Accuracy')
plt.legend()
plt.figure()

plt.plot(epochs, loss2, 'b', label = 'Train loss')
plt.plot(epochs, val_loss, 'r', label = 'Test loss')
plt.legend()
plt.show()
```

# Model with initializer and regularizer

Techniques to overcome overfitting/underfitting

# Kernal Regularisation technique:
Adding weight penalities factors to Network layer in weight values to allow the network converge optimally.
 
**Kernal initialisers**:

Used to statistically initialise weights in the model. Generate weights and distribute them which is used as starter weights.

**Regularisation**: A set of tehcniques that prevent overfitting in NN which improves the accuracy of DL model when facing new data from the problem domain.

```python
from keras import initializers
from keras import regularizers
```

```python
# Build a linear stack of layers with Sequential Model
# Basically, bring back the previous code and run.
# But, you may add more kernel initializers and regularisers in 2nd layer, 3rd layer and so on. This is to check whether there's improvements in your model.

# Number of neurals
L1 = 100 # First hidden layer neurons
L2 = 100 # Second hidden layer neurons

# This means parameters in input model.
input = 30

# Activation functions
activation_function = 'relu'
activation_function2 = 'relu'
output_activation_function = 'sigmoid'

# Kernel initializers
kernel_initializer1 = 'he_normal'

# Number of outputs
# You need to know your columns, for example, you only made a column "Target", so here only one value.
no_of_output = 1

# Initialize object
model = Sequential()

# First hidden layer
model.add(Dense(L1, input_dim=input, activation=activation_function, kernel_initializer=kernel_initializer1, kernel_regularizer = regularizers.l2(0.01)))

# Second hidden layer
model.add(Dense(L2, activation=activation_function2))

# Output layer
model.add(Dense(no_of_output, activation=output_activation_function))
```

```python
# Bring back the previous code and run. No changes in this part.
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
# Bring back the previous code and run. No changes in this part.
history = model.fit(X_train, Y_train, epochs = 10, batch_size = 100, validation_data = (X_test, Y_test))
```

```python
# Bring back the previous code and run. No changes in this part.
# Visualisations

import matplotlib.pyplot as plt

acc2 = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss2 = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1,len(acc2)+1)

# 'b' and 'r' are blue and red color

plt.plot(epochs, acc2, 'b', label='Train Accuracy')
plt.plot(epochs, val_acc, 'r', label='Test Accuracy')
plt.legend()
plt.figure()

plt.plot(epochs, loss2, 'b', label = 'Train loss')
plt.plot(epochs, val_loss, 'r', label = 'Test loss')
plt.legend()
plt.show()
```
